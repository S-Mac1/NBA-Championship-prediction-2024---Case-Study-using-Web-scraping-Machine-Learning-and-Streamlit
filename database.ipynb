```python
# Import der Module für die Erstellung des leeren DataFrames
import pandas as pd

df = pd.DataFrame(columns = ['season', 'team', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'OFG', 'OFGA', 'OFG%', 'O3P', 'O3PA', 'O3P%', 'O2P', 'O2PA', 'O2P%', 'OFT', 'OFTA', 'OFT%', 'OORB', 'ODRB', 'OTRB', 'OAST', 'OSTL', 'OBLK', 'OTOV', 'OPF', 'OPTS', 'champion'])
```

```python
df
```

```python
# Liste erstellen für alle NBA Saisons von 2024 bis 1947
seasons = list(range(2024, 1946, -1))
```

```python
print(seasons)
```

```python
# Import der Module für Webscraping
import requests
from bs4 import BeautifulSoup

# Webscraping der Saisonstatistiken von 2024 bis 1947
for season in seasons:

    if season in [1949, 1948, 1947]:

        # URL der Webseite mit den Per Game Stats für BAA
        url = f"https://www.basketball-reference.com/leagues/BAA_{season}.html#all_per_game_team-opponent"

    else:
        
        # URL der Webseite mit den Per Game Stats für NBA
        url = f"https://www.basketball-reference.com/leagues/NBA_{season}.html#all_per_game_team-opponent"
   

    # Die Seite herunterladen
    response = requests.get(url)

    # BeautifulSoup verwenden, um den HTML-Inhalt zu analysieren
    soup = BeautifulSoup(response.content, "html.parser")

    # Die Tabelle mit den Per Game Stats (Offensive) finden
    table = soup.find("table", {"id": "per_game-team"})

    if table is not None:

    # Die Zeilen der Tabelle durchgehen und die Daten extrahieren
        for row in table.find_all("tr")[1:-1]:
            columns = row.find_all("td")
            team_name = columns[0].text
            team_name = team_name.replace("*", "")
            field_goals = columns[3].text
            field_goal_attempts = columns[4].text
            field_goal_percentage = columns[5].text
            three_point_field_goals = columns[6].text
            three_point_field_goal_attempts = columns[7].text
            three_point_field_goal_percentage = columns[8].text
            two_point_field_goals = columns[9].text
            two_point_field_goal_attempts = columns[10].text
            two_point_field_goal_percentage = columns[11].text
            free_throws = columns[12].text
            free_throw_attempts = columns[13].text
            free_throw_percentage = columns[14].text
            offensive_rebounds = columns[15].text
            defensive_rebounds = columns[16].text
            total_rebounds = columns[17].text
            assists = columns[18].text
            steals = columns[19].text
            blocks = columns[20].text
            turnovers = columns[21].text
            fouls = columns[22].text
            points = columns[23].text

            df = pd.concat([df, pd.DataFrame({
                        'season': [season],
                        'team': [team_name],
                        'FG': [field_goals],
                        'FGA': [field_goal_attempts],
                        'FG%': [field_goal_percentage],
                        '3P': [three_point_field_goals],
                        '3PA': [three_point_field_goal_attempts],
                        '3P%': [three_point_field_goal_percentage],
                        '2P': [two_point_field_goals],
                        '2PA': [two_point_field_goal_attempts],
                        '2P%': [two_point_field_goal_percentage],
                        'FT': [free_throws],
                        'FTA': [free_throw_attempts],
                        'FT%': [free_throw_percentage],
                        'ORB': [offensive_rebounds],
                        'DRB': [defensive_rebounds],
                        'TRB': [total_rebounds],
                        'AST': [assists],
                        'STL': [steals],
                        'BLK': [blocks],
                        'TOV': [turnovers],
                        'PF': [fouls],
                        'PTS': [points]
                    })], ignore_index=True)
            
    else:
        print("Die Tabelle wurde nicht gefunden.")

        
        # Die Tabelle mit den Per Game Stats (Defensive) finden
    table_opponent = soup.find("table", {"id": "per_game-opponent"})

    if table_opponent is not None:

        # Die Zeilen der Tabelle durchgehen und die Daten extrahieren
        for row in table_opponent.find_all("tr")[1:-1]:
            columns = row.find_all("td")
            team_name = columns[0].text
            team_name = team_name.replace("*", "")
            o_field_goals = columns[3].text
            o_field_goal_attempts = columns[4].text
            o_field_goal_percentage = columns[5].text
            o_three_point_field_goals = columns[6].text
            o_three_point_field_goal_attempts = columns[7].text
            o_three_point_field_goal_percentage = columns[8].text
            o_two_point_field_goals = columns[9].text
            o_two_point_field_goal_attempts = columns[10].text
            o_two_point_field_goal_percentage = columns[11].text
            o_free_throws = columns[12].text
            o_free_throw_attempts = columns[13].text
            o_free_throw_percentage = columns[14].text
            o_offensive_rebounds = columns[15].text
            o_defensive_rebounds = columns[16].text
            o_total_rebounds = columns[17].text
            o_assists = columns[18].text
            o_steals = columns[19].text
            o_blocks = columns[20].text
            o_turnovers = columns[21].text
            o_fouls = columns[22].text
            o_punkte = columns[23].text

            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFG'] = o_field_goals
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFGA'] = o_field_goal_attempts
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFG%'] = o_field_goal_percentage
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O3P'] = o_three_point_field_goals
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O3PA'] = o_three_point_field_goal_attempts
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O3P%'] = o_three_point_field_goal_percentage
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O2P'] = o_two_point_field_goals
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O2PA'] = o_two_point_field_goal_attempts
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'O2P%'] = o_two_point_field_goal_percentage
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFT'] = o_free_throws
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFTA'] = o_free_throw_attempts
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OFT%'] = o_free_throw_percentage
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OORB'] = o_offensive_rebounds
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'ODRB'] = o_defensive_rebounds
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OTRB'] = o_total_rebounds
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OAST'] = o_assists
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OSTL'] = o_steals
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OBLK'] = o_blocks
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OTOV'] = o_turnovers
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OPF'] = o_fouls
            df.loc[(df['team'] == team_name) & (df['season'] == season), 'OPTS'] = o_punkte

    else:
        print("Die Tabelle wurde nicht gefunden.")
```

```python
# Der fertige DataFrame mit allen Statistiken der NBA-Teams von 2024 bis 1947
print(df)
```
